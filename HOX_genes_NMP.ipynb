## Import packages

```python
import pandas as pd
import numpy as np
import scanpy as sc
import scanpy.external as sce
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
import math
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
sc.settings.verbosity = 0  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.set_figure_params(scanpy=True, dpi=100, dpi_save=500, frameon=False, vector_friendly=True, fontsize=14, figsize=None, color_map=None, format='pdf', facecolor=None, transparent=True)
```

## Load

```python
segmentoid = sc.read('data/segmentoid_processed.h5ad')
```

```python
segmentoid = segmentoid[segmentoid.obs['leiden'].isin(['NMP']),:]
```

## Scraping

```python
HOXA = [i for i in segmentoid.var_names if i.startswith('HOXA')]
HOXA = [x for x in HOXA if "AS" not in x]

HOXB = [i for i in segmentoid.var_names if i.startswith('HOXB')]
HOXB = [x for x in HOXB if "AS" not in x]

HOXC = [i for i in segmentoid.var_names if i.startswith('HOXC')]
HOXC = [x for x in HOXC if "AS" not in x]

HOXD = [i for i in segmentoid.var_names if i.startswith('HOXD')]
HOXD = [x for x in HOXD if "AS" not in x]
```
